# Sampling Theorem
## Import the Librarys

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft
```

## Create fuction to Caluclate furrier Transform

```python
def calc_fft(xAxis,yAxis):
    n = np.size(xAxis) # get total size of xAxis
    x = fft(yAxis)    # forrier trasnsform
    X_m = (2/n)*abs(x)
    return X_m
```

## Create fuction to plot Data
### Fuction takes following arguments
- **CT** : True for Continue / False for Discreate
- **xAxis** : Value of x axis in array format
- **yAxis** : Value of y axis in array format

```python
def plot_data(CT,xAxis,yAxis):
    plt.subplot(2,1,1)
    if CT:
        plt.plot(xAxis,yAxis)  # used to plot Contines time series
    else:
        plt.stem(xAxis,yAxis,use_line_collection = True) # used to plot dicrete samples
    plt.title("Sinusidal Signal in time domain")
    plt.xlabel("time")
    plt.ylabel("Amplitude")
    plt.show()

    plt.subplot(2,1,2)
    plt.title("Sinusidal Signal in Freq domain")
    plt.xlabel("time")
    plt.ylabel("Frequency")
    if CT:
        plt.plot(calc_fft(xAxis,yAxis))
    else:
        plt.stem(calc_fft(xAxis,yAxis),use_line_collection = True)
    plt.show()
```

## Signal without Sampling

```python
pi = 3.14
FS = 200
f1 = 10
```

```python
t = np.arange(0,1,(1/FS))
sine = np.sin(2*pi*f1*t)
```

```python
plot_data(True,t,sine)
```

## Sampling the Signal

```python
Fs = 40 # Sampling Frequency
f1 = 10 # Signal 1 frequency
f2 = 50 # Signal 1 frequency
n = np.arange(0,Fs,1)
```

```python
x1 = np.sin(2*pi*n*(f1/Fs))
```

### Signal 1 are smpled with 2f <= Fs

```python
plot_data(True,n,x1)
```
